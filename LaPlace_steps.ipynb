```python
x = np.arange(0, 6, 0.01)
fx=np.exp(-x)*np.sin(x)
new_x = np.arange(-3, 3, 0.01)
iw = 0+new_x*1j

plt.figure(figsize=(9, 3), dpi=300)
plt.plot(x, fx)
plt.title(r"$f(t) = e^{-t} sin(t)$", size = 10)
plt.xlabel('t')
plt.ylabel('f(t)')
```

'$FT = \int_{0}^{\infty} e^{-t} sin(t) e^{-iwt}dt$...............Eq(1)'

'$FT = \frac{1}{1+(1+iw)^2}$ '

```python
FT = np.divide(1,1+np.power(1+iw,2))
magnitude = abs(FT)
```

```python

plt.figure(figsize=(9, 3), dpi=300)
plt.plot(new_x, magnitude)
plt.title(r"$|FT| = |\frac{1}{1+(1+iw)^2}|$", size = 10)
plt.xlabel('w')
plt.ylabel('Magnitude')
```

```python
FT_apf = np.divide(1,1+np.power(0.5+iw,2))
magnitude_apf = abs(FT_apf)
```

'Now say you want to modify the function to $  f(x) = e^{-t} e^{-0.5t}sin(t)$  ......... Eq(1)'

'The integral will then be $\int_{0}^{\infty} e^{-t} sin(t) e^{-0.5t} e^{-iwt}dt$ ............ Eq(2)'

'Basically what you are doing is obtainning the fourier transform of $e^{-t} sin(t) e^{-0.5t}$'

'which is also equivalent to replacing the variable iw for $\alpha + iw$, where alpha in Eq(1) is  0 and now is -0.5'

'The result of Eq(2) is $FT = \frac{1}{1+(0.5+iw)^2}$, which yields to'

```python
plt.figure(figsize=(9, 3), dpi=300)
plt.plot(new_x, magnitude_apf)
plt.title(r"$|FT| = |\frac{1}{1+(\alpha+iw)^2}| \quad \alpha = 0.5 $", size = 10)
plt.xlabel('w')
plt.ylabel('Magnitude')
```

'Now if we keep varying alpha, in this case approximating it to -1'

'$\alpha = -0.6, \alpha = -0.7,  \alpha = -0.8,  \alpha= -0.9,  \alpha = -0.91,  \alpha = -0.92,  \alpha = -0.93$.... '

'$ \rightarrow |FT_{\alpha=-0.6}| = |\frac{1}{1+(0.4+iw)^2}|$'

'$ \rightarrow |FT_{\alpha=-0.7}| = |\frac{1}{1+(0.3+iw)^2}|$'

'$ \rightarrow |FT_{\alpha=-0.8}| = |\frac{1}{1+(0.2+iw)^2}|$'

'$ \rightarrow |FT_{\alpha=-0.9}| = |\frac{1}{1+(0.1+iw)^2}|$'

'$ \rightarrow |FT_{\alpha=-0.91}| = |\frac{1}{1+(0.09+iw)^2}|$'

'$ \rightarrow |FT_{\alpha=-0.92}| = |\frac{1}{1+(0.08+iw)^2}|$'

'$ \rightarrow |FT_{\alpha=-0.93}| = |\frac{1}{1+(0.07+iw)^2}|$'

```python
FT_06 = np.divide(1,1+np.power(0.4+iw,2))
FT_07 = np.divide(1,1+np.power(0.3+iw,2))
FT_08 = np.divide(1,1+np.power(0.2+iw,2))
FT_086 = np.divide(1,1+np.power(0.14+iw,2))
FT_09 = np.divide(1,1+np.power(0.1+iw,2))
FT_091 = np.divide(1,1+np.power(0.09+iw,2))
FT_092 = np.divide(1,1+np.power(0.08+iw,2))
FT_093 = np.divide(1,1+np.power(0.07+iw,2))

magnitude_06 = abs(FT_06)
magnitude_07 = abs(FT_07)
magnitude_08 = abs(FT_08)
magnitude_086 = abs(FT_086)
magnitude_09 = abs(FT_09)
magnitude_091 = abs(FT_091)
magnitude_092 = abs(FT_092)
magnitude_093 = abs(FT_093)
```

```python
plt.figure(figsize=(9, 3), dpi=300)
plt.plot(new_x, magnitude_06, label = r"$\alpha = 0.6$")
plt.plot(new_x, magnitude_07,label = r"$\alpha = 0.7$")
plt.plot(new_x, magnitude_08,label = r"$\alpha = 0.8$")
plt.plot(new_x, magnitude_09,label = r"$\alpha = 0.9$")
plt.plot(new_x, magnitude_091,label = r"$\alpha = 0.91$")
plt.plot(new_x, magnitude_092,label = r"$\alpha = 0.92$")
plt.plot(new_x, magnitude_093,label = r"$\alpha = 0.93$")



plt.title(r"$|FT| = |\frac{1}{1+(\alpha+iw)^2}| \quad$", size = 10)
plt.xlabel('w')
plt.ylabel('Magnitude')
plt.legend()

```
